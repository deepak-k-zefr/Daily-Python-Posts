```python
## Monty Hall Problem
```

```python
right=np.random.choice(3,100000)
my_choice=np.random.choice(3,100000)
```

```python
monty_choice=[]
for a,b in zip(right,my_choice):
    if 0 not in list([a,b]):
        monty_choice.append(0)
    elif 1 not in list([a,b]):
        monty_choice.append(1)
    else:
        monty_choice.append(2)
```

```python
same=[]
for x in range(10000):
    right=np.random.choice(3,x)
    my_choice=np.random.choice(3,x)
    same.append((my_choice==right).sum()/x)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=[30,15])
plt.plot(same)
```
