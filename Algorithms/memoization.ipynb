```python
import time


## Without Memoization
def expensive_func(num):
    print("Computing {}....".format(num))
    time.sleep(1)
    return num*num


result = expensive_func(4)
print (result)

result = expensive_func(10)
print (result)

result = expensive_func(4)
print (result)

result = expensive_func(10)
print (result)
```

```python
## With Memoization:
ef_cache={}
def expensive_func(num):
    if num in ef_cache:
        return ef_cache[num]
    
    print("Computing {}....".format(num))
    ef_cache[num]=num*num
    time.sleep(1)
    return num*num

result = expensive_func(4)
print (result)

result = expensive_func(10)
print (result)

result = expensive_func(4)
print (result)

result = expensive_func(10)
print (result)
```

```python
cache={}
def fibo(num):
    if num==1 or num==0:
        return 1
    elif num in cache:
        return cache[num]
    else:
        cache[num]=fibo(num-1)+fibo(num-2)
        return cache[num]
```
